```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv("../data/processed/orders_clean.csv")

order_items = pd.read_csv("../data/processed/order_items_clean.csv")

products = pd.read_csv("../data/processed/products_clean.csv")

payments = pd.read_csv("../data/processed/payments_clean.csv")

reviews = pd.read_csv("../data/processed/reviews_clean.csv")

customers = pd.read_csv("../data/raw/olist_customers_dataset.csv")
```

```python
df = orders.merge(
    customers,
    on="customer_id",
    how="left"
)
```

```python
df.shape
```

```python
df = df.merge(
    order_items,
    on="order_id",
    how="left"
)
```

```python
df.shape
```

```python
df = df.merge(
    products,
    on="product_id",
    how="left"
)
```

```python
df = df.merge(
    payments,
    on="order_id",
    how="left"
)
```

```python
df = df.merge(
    reviews,
    on="order_id",
    how="left"
)
```

```python
df.info()
```

```python
df.head()
```

```python
df.shape
```

```python
df["product_category_name"] = (
    df["product_category_name"]
    .fillna("Unknown")
)
```

```python
df["payment_type"] = df["payment_type"].fillna("Unknown")

df["review_comment_title"] = df["review_comment_title"].fillna("No Title")

df["review_comment_message"] = df["review_comment_message"].fillna("No Review")
```

```python
print("Shape:", df.shape)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
df = df.sort_values("order_purchase_timestamp").reset_index(drop=True)
```

```python
df.to_csv("../data/processed/ecommerce_master_dataset.csv", index=False)
```

```python
df.info()
```

```python
date_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date",
    "shipping_limit_date",
    "review_creation_date",
    "review_answer_timestamp"
]

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors="coerce")
```

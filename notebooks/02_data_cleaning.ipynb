```python
date_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_columns:
    orders_clean[col] = pd.to_datetime(orders_clean[col])
```

```python
orders_clean.dtypes
```

### Date Conversion

The order-related timestamp columns were converted from string (`object`) to `datetime64[ns]` to enable time-based analysis such as:

- Monthly sales trends
- Delivery duration analysis
- Order processing time
- Customer purchase patterns
- Time-series visualizations

This conversion improves analytical accuracy and allows Pandas to perform efficient date calculations.

### Handling Missing Timestamp Values

The timestamp columns (`order_approved_at`, `order_delivered_carrier_date`, and `order_delivered_customer_date`) contain missing values.

These values were **not removed or imputed**, as they correspond to orders that were cancelled, unavailable, or not delivered. The missing timestamps accurately reflect the order lifecycle and therefore were retained to preserve business context.

```python
orders_clean[orders_clean["order_delivered_customer_date"].isnull()]["order_status"].value_counts()
```

```python
orders_clean[orders_clean["order_delivered_carrier_date"].isnull()]["order_status"].value_counts()
```

```python
orders_clean[orders_clean["order_approved_at"].isnull()]["order_status"].value_counts()
```

### Cleaning Decision - Orders Dataset

The missing values in the timestamp columns were investigated before applying any cleaning operations.

Observations:
- Missing delivery timestamps mainly belong to cancelled, unavailable, shipped, invoiced, and processing orders.
- Missing approval timestamps mainly belong to cancelled orders.

Decision:
These missing values represent the normal order lifecycle rather than data quality issues. Therefore, no rows were removed and no values were imputed.

```python
orders_clean["delivery_days"] = (
    orders_clean["order_delivered_customer_date"]
    - orders_clean["order_purchase_timestamp"]
).dt.days
```

```python
orders_clean["approval_time_hours"] = (
    orders_clean["order_approved_at"]
    - orders_clean["order_purchase_timestamp"]
).dt.total_seconds() / 3600
```

```python
orders_clean["delivery_delay_days"] = (
    orders_clean["order_delivered_customer_date"]
    - orders_clean["order_estimated_delivery_date"]
).dt.days
```

```python
orders_clean["order_month"] = (
    orders_clean["order_purchase_timestamp"].dt.to_period("M").astype(str)
)
```

```python
orders_clean["order_year"] = (
    orders_clean["order_purchase_timestamp"].dt.year
)
```

```python
orders_clean["purchase_weekday"] = (
    orders_clean["order_purchase_timestamp"].dt.day_name()
)
```

What makes us build them?

Think about the manager asking the following question:

"During which month did we have the largest number of orders?"

And there is no need to recalculate the month each time since order_month has been built before.

Or:

"What is the average number of days required for delivery?"

There is delivery_days that has been built before.

Such features are called engineered features.
